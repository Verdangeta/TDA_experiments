```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Загружаем изображение
image = cv2.imread('pic_1.png')

# Конвертируем в оттенки серого
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Бинаризуем изображение
threshold, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Находим контуры на изображении
contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Выбираем самый большой контур (предполагаем, что это график)
largest_contour = max(contours, key=cv2.contourArea)

# Получаем координаты точек контура
x, y = zip(*largest_contour.reshape(-1, 2))


# Пример масштабирования и калибровки
xmax = 1
xmin = 0

ymax = 2.5
ymin = 1.5
x = (np.array(x) - min(x)) / (max(x) - min(x)) * (xmax - xmin) + xmin
y = (1 - (np.array(y) - min(y)) / (max(y) - min(y))) * (ymax - ymin) + ymin

# Строим график
plt.plot(x, y, linewidth=2.0)
plt.ylabel('Cross-entropy')
plt.ylim((1.4,2.6))
plt.show()
```
